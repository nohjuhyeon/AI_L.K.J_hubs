## 지역별 관광지 카테고리 방문자수 (비율)

### 지역별
- 인천 : 자연관광지, 휴양관광지, 역사관광지
- 부산 : 자연관광지, 건축/조형물, 체험관광지
- 서울 : 체험관광지, 역사관광지, 휴양관광지
- 제주 : 자연관광지, 휴양관광지, 문화시설
- 강원 : 자연관광지, 체험관광지, 역사관광지

```python
incheon_visitors = df_attraction.query(condition_incheon).groupby('KTO카테고리중분류명')['방문자수'].sum().sort_values(ascending=False)
sum_incheon = incheon_visitors.sum()
percentages_incheon = (incheon_visitors / sum_incheon) * 100

plt.pie(percentages_incheon, labels=incheon_visitors.index, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```
